```python
total_bookings = hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).count()
sum_cancels = hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).sum()
print(total_bookings)
print(sum_cancels)
```

```python
sum_cancels / total_bookings 
```

```python
num_cancelations = hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).sum() 
num_bookings = hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).count() 

# Percent of Bookings that did not cancel
percent_cancelations = 1 - num_cancelations / num_bookings
```

```python
percent_cancelations
```

```python
percent_cancelations.reset_index(inplace = True)


percent_cancelations.rename(columns = \
                            {'arrival_date_year':'Year',
                             'is_canceled': 'pct_bookings_used'},
                           inplace = True)

percent_cancelations['pct_bookings_used'] = percent_cancelations['pct_bookings_used'] * 100
```

```python
percent_cancelations.plot(kind = 'bar',
                          legend = False,
                          x = 'Year',
                          y = 'pct_bookings_used')
```
