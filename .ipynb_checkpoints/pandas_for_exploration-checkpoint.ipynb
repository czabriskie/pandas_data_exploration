### Now let's answer some of our previous questions 

1. What is the proportion of cancels to non-cancels?

There are several ways to subset a dataframe
- `cancels = hotel_data.is_canceled`
- `cancels = hotel_data['is_canceled']`
- `cancels = hotel_data.loc[:, 'is_canceled']`
- `cancels = hotel_data.iloc[:, 1]`

```python
cancels = hotel_data.is_canceled
```

Now we can see how many cancels there are to non cancels

```python
cancels.value_counts()
```

To get a proportion, we just need to divide each column by the total.

```python
cancels.value_counts() / len(cancels)
```

Okay, so this is cool and all, but I'm assuming this has changed every year, what would be cool is if we had a year over year comparison of this

```python
hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).count()
```

This also brought up something new, why do we see a decrease of bookings from 2016 to 2017?

```python
num_cancelations = hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).sum() 
num_bookings = hotel_data[['arrival_date_year', 'is_canceled']].groupby(['arrival_date_year']).count() 

# Percent of Bookings that did not cancel
percent_cancelations = 1 - num_cancelations / num_bookings

percent_cancelations
```

Looks like the last year was our worst year in regards to keeping guest bookings booked. That combined with the decrease in overall bookings year over year is a pretty big red flag that something isn't working right. However, could this just be due to random chance?

```python
percent_cancelations.reset_index(inplace = True)
percent_cancelations.rename(columns = \
                            {'arrival_date_year':'Year',
                             'is_canceled': 'pct_bookings_used'},
                           inplace = True)

percent_cancelations['pct_bookings_used'] = percent_cancelations['pct_bookings_used'] * 100
```

```python
percent_cancelations.plot(kind = 'bar',
                          legend = False,
                          x = 'Year',
                          y = 'pct_bookings_used')
```

The above plot could do with some work, but more concerning is that it doesn't really show the story we want to tell, that we have decreased from last year. What other plot would be good here?



Perhaps a YoY plot line plot by month would show this a bit better

```python
# Create a new column that uses a month number instead of month name for easier plotting
conditions = [
    (hotel_data['arrival_date_month'] == 'January'),
    (hotel_data['arrival_date_month'] == 'February'),
    (hotel_data['arrival_date_month'] == 'March'),
    (hotel_data['arrival_date_month'] == 'April'),
    (hotel_data['arrival_date_month'] == 'May'),
    (hotel_data['arrival_date_month'] == 'June'),
    (hotel_data['arrival_date_month'] == 'July'),
    (hotel_data['arrival_date_month'] == 'August'),
    (hotel_data['arrival_date_month'] == 'September'),
    (hotel_data['arrival_date_month'] == 'October'),
    (hotel_data['arrival_date_month'] == 'November'),
    (hotel_data['arrival_date_month'] == 'December'),
]
choices = [x for x in range(1, 13)]
hotel_data['arrival_date_month_num'] = np.select(conditions, choices)
```

```python
# Get the cancelations each year per month
num_month_cancelations = hotel_data[['arrival_date_year', 'arrival_date_month_num', 'is_canceled']].groupby(['arrival_date_year', 'arrival_date_month_num']).sum() 
num_month_bookings = hotel_data[['arrival_date_year', 'arrival_date_month_num', 'is_canceled']].groupby(['arrival_date_year', 'arrival_date_month_num']).count()

# reset index so we can get years for each column
num_month_cancelations.reset_index(inplace = True)
num_month_bookings.reset_index(inplace = True)
```

```python
num_month_bookings
```

Seasonality could be a huge factor here in why there are more cancelations in one year over another. It looks like we don't have data for a lot of the months in 2015 and 2017

```python
num_month_bookings['pct_bookings_used'] = num_month_cancelations.is_canceled / num_month_bookings.is_canceled * 100 
```

```python
num_month_bookings
```

```python
# create new index for plotting 
# https://medium.com/@sciencelee/making-plots-with-the-pandas-groupby-ac492941af28
vals = num_month_bookings.groupby(['arrival_date_month_num', 'arrival_date_year'])['pct_bookings_used'].sum().unstack()
vals.replace(0, np.nan, inplace = True)
```

```python
vals
```

```python
vals.plot(kind = 'line')
```

Looking at the above plot, we can see that if we look at the data year over year, we are doing much better than we were in 2016 as far as keeping bookings. However, it is interesting how far we declined in 2015 over just a few months.

The last question I had concerning booking was the difference between repeat guests and non-repeat guests in how much they cancel?

```python
# Create new column that indicates if a guest is a repeat customer
total_repeats = hotel_data[['is_repeated_guest', 'is_canceled']].groupby('is_repeated_guest').count()
total_canceled_repeats = hotel_data[['is_repeated_guest', 'is_canceled']].groupby('is_repeated_guest').sum()

print(total_repeats)
print(total_canceled_repeats)
```

Not a lot of our bookings come from repeat clients. This seems odd are they not having a good experience, or is it because we only have a few years of data?

```python
total_repeats = (1 - total_canceled_repeats / total_repeats) * 100
total_repeats.reset_index(inplace = True)
total_repeats.rename(columns = {'is_canceled': 'pct_kept_booking'},
                    inplace = True)

total_repeats
```

It is interesting to see that the percentage of the repeat customers that keep their booking is significantly higher. This would be a good one to show in a plot I think

```python
total_repeats['pct_kept_booking'].plot(kind = 'bar')
```
